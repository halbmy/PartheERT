```python
from pygimli.physics import ert
```

```python
deldata = [[1243, 1248, 1289, 1294, 1302], # a
           [529, 689, 649, 1087], # b
           [1004, ], # c
           [43, 45], # d
           [547,], # e
           [],  # beta - a
           [642, 650, 723, 770, 795, 466, 853],  # b
           []]  # e
```

```python
DATA = [ert.load(f"Parthe2023III{p}wen.dat") for p in "abcde"]
DATA.extend([ert.load(f"Parthe2023III{p}wb.dat") for p in "abe"])
```

```python
alldata = ert.DataContainer()
for i, data in enumerate(DATA):
    for dd in deldata[i]:
        data.markInvalid(dd-1)
    data.removeInvalid()
    alldata.add(data)

ert.show(alldata)
```

```python
alldata["err"] = ert.estimateError(alldata, relativeError=0.02)
alldata.save("T3-alldata.dat", "a b m n k rhoa err")
```

```python
import numpy as np
allxz = np.genfromtxt("parthe2023T3-topo.txt")
for i, xz in enumerate(allxz):
    alldata.setSensor(i, xz)

alldata.save("T3-alldata.ohm", "a b m n u i")
```
