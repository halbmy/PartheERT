```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y, x, z = np.genfromtxt("Leipzig-JUNI2023T3.TXT", delimiter="\t",
                        usecols=(1, 2, 3), unpack=True)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, ".-")
ax.set_aspect(1.0)
ax.grid()
```

```python
dh = np.sqrt(np.diff(x)**2 + np.diff(y)**2)
print(min(dh), max(dh), np.median(dh))
np.argmin(dh)
```

```python
from pygimli.frameworks import harmfitNative as harmfit
h = np.hstack([0., np.cumsum(dh)])
hC = np.arange(0, 140.6, 0.5)
zC, _ = harmfit(z, h, nc=20, xc=hC)
fig, ax = plt.subplots()
ax.plot(h, z, ".-")
ax.plot(hC, zC, "r-")
```

```python
np.savetxt("parthe2023T3-topo.txt", np.column_stack([hC, zC]), fmt="%.2f")
```
